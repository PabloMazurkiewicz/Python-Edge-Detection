Model 1 (Core)

```python
import cv2 

img = cv2.imread("Car1.jpg") # Read image 
img_copy = img.copy() # Create a copy of the image

# Setting parameter values 
t_lower = 0 # Lower Threshold 
t_upper = 150 # Upper threshold 
aperturesize = 5 # Aperture Size
L2Gradient = True # Boolean

# Applying the Canny Edge filter 
edge = cv2.Canny(img_copy, t_lower, t_upper) 

cv2.imshow("Actual", img) 
cv2.imshow("Processed", edge) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

Model 2 (Showing Various Subplots)

```python
import cv2 as cv
from matplotlib import pyplot as plt
 
img = cv.imread("Car1.jpg")
img_copy = img.copy() # Create a copy of the image
edges = cv.Canny(img_copy, 75, 100)

plt.figure(figsize=(16,9))
plt.subplot(221), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(edges)
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

Contour Detection Model

```python
import cv2
import numpy as np

# Let's load a simple image with 3 black squares
image = cv2.imread('Car1.jpg')
image_copy = image.copy() # Create a copy of the image
cv2.waitKey(0)

# Grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

# Find Canny edges
edged = cv2.Canny(gray, 30, 200)
cv2.waitKey(0)

# Finding Contours
# Use a copy of the image e.g. edged.copy()
# since findContours alters the image
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

cv2.imshow('Canny Edges After Contouring', edged)
cv2.waitKey(0)

print("Number of Contours found = " + str(len(contours)))

# Draw all contours
# -1 signifies drawing all contours
cv2.drawContours(image_copy, contours, -1, (0, 255, 0), 3)

cv2.imshow('Contours', image_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Enhanced Version

```python
import cv2 as cv
from matplotlib import pyplot as plt
 
image = cv.imread("Car1.jpg", cv.IMREAD_COLOR)
image_copy = image.copy() # Create a copy of the image
edges = cv.Canny(image_copy, 75, 100, L2gradient=True, apertureSize=3)
BLedges = cv.Canny(image_copy, 75, 100, cv.IMREAD_GRAYSCALE)
contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)

plt.figure(figsize=(16,9))
plt.subplot(221), plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(edges, cmap='gray')
plt.title('B&L Edge Image'), plt.xticks([]), plt.yticks([])

cv.drawContours(image_copy, contours, -1, (50, 205, 50), 2)

plt.subplot(224), plt.imshow(cv.cvtColor(image_copy, cv.COLOR_BGR2RGB))
plt.title('Contour Image'), plt.xticks([]), plt.yticks([])

plt.show()

print("Number of Contours found = " + str(len(contours)))

plt.imsave('Contour_result.png', image_copy, cmap="gray", format='png')
plt.imsave('Edge_result.png', edges, cmap="gray", format='png')
```
